```python
import pandas as pd

df = pd.read_csv('data/train.csv')

df
```

```python


from sklearn import preprocessing

le = preprocessing.LabelEncoder()

df['Sex'] = le.fit_transform(df['Sex'])

df['Embarked'] = le.fit_transform(df['Embarked'])

df['Cabin'] = le.fit_transform(df['Cabin'])

age_mean = df['Age'].mean()

df['Age'].fillna(age_mean, inplace=True)

df
```

```python
X = df.drop(['Name', 'Ticket', 'Survived'], axis=1)
y = df['Survived']

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

```

```python

from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(random_state=0).fit(X_train, y_train)

lr_y_pred = lr.predict(X_test)

from sklearn.tree import DecisionTreeClassifier

df = DecisionTreeClassifier(random_state=0).fit(X_train, y_train)

df_y_pred = df.predict(X_test)

from sklearn.metrics import accuracy_score

lr_acc = accuracy_score(y_test, lr_y_pred)

df_acc = accuracy_score(y_test, df_y_pred)

from sklearn.metrics import recall_score

lr_rc = recall_score(y_test, lr_y_pred, average='macro')

df_rc = recall_score(y_test, df_y_pred, average='macro')

print(lr_acc, df_acc, lr_rc, df_rc)

```

```python
from sklearn.model_selection import GridSearchCV

lr_param_grid = {'C': [0.1, 1, 10, 100, 1000], 'penalty': ['l1', 'l2']}

df_param_grid = {'criterion': ['gini', 'entropy'], 'max_depth': [4, 5, 6, 7, 8]}

lr_cv = GridSearchCV(estimator=lr, param_grid=lr_param_grid, cv=5, scoring='accuracy')

df_cv = GridSearchCV(estimator=df, param_grid=df_param_grid, cv=5, scoring='accuracy')

lr_cv.fit(X_train, y_train)

df_cv.fit(X_train, y_train)

lr_best_Model = lr_cv.best_estimator_

df_best_Model = df_cv.best_estimator_

print(lr_best_Model.coef_)

print(df_best_Model.feature_importances_)

lr_best_pre = lr_best_Model.predict(X_test)

df_best_pre = df_best_Model.predict(X_test)
```

```python
import joblib

joblib.dump(lr_best_Model, 'lr_best_Model.pkl')
joblib.dump(df_best_Model, 'df_best_Model.pkl')
```
